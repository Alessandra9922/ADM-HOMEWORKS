### Exercise n° 5 - Find a String
##### Reference: https://www.hackerrank.com/challenges/find-a-string/problem?isFullScreen=true
##### This exercise is focused on string and substring. The task is to print the number of times that the substrings occurs in the string

```python
#######################################################################
#                                                                     #
# MY FUNCTION: In a single line of code I directly return the number  #
#              of times that a substring in in a string using list    #
#              list comprehension, adding 1 each time I find the      #
#              the substring number of letters in string. I use the   #
#              indexing of a string                                   # 
#                                                                     #    
#######################################################################

def count_substring(string, sub_string):
    return sum([1 for i in range(len(string)-len(sub_string)+1) if string[i:i+len(sub_string)] == sub_string])

####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

### Exercise n° 7 - Text Alignment
##### Reference: https://www.hackerrank.com/challenges/text-alignment/problem?isFullScreen=true
##### This exercise is focused on printing in a particolar layout. The task is generating the HackerRank Logo of variable thickness using rjust, ljust or center.

```python
###################################################################
#                                                                 #    
# MY CODE: I use .rjust, .ljust and .center in a right way to     #
#          print the logo                                         #
#                                                                 #        
###################################################################

thickness = int(input()) #This must be an odd number
c='H'
#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))
#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))
#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    
#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    
#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```

### Exercise n° 11 - Alphabet Rangoli
##### Reference: https://www.hackerrank.com/challenges/alphabet-rangoli/problem?isFullScreen=true
##### This exercise is focused to design rangoli

```python
########################################################################
#                                                                      #    
# MY FUNCTION: I print Rangoli using the English alphabet, centering   #
#              and joining the letters with the '-' character.         #
#              The list 'l' contains all the rows of the Rangoli.      #
#              The 'print' line combines and prints the rows in reverse# 
#              order followed by the normal order to form the complete #
#              pattern                                                 #
#                                                                      #        
########################################################################

def print_rangoli(size):
    import string
    rangoli = string.ascii_lowercase
    l = []
    for i in range(size):
        s = "-".join(rangoli[i:size])
        l.append((s[::-1]+s[1:]).center(4*size-3, "-")) 
    print('\n'.join(l[:0:-1]+l))
    return

####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

### Exercise n° 13 - The Minion Game
##### Reference: https://www.hackerrank.com/challenges/the-minion-game/problem?isFullScreen=true
##### This exercise is focused on extrxting substrings and analize strings. The task is to determine the winner of the game and their score, using command associated to the string

```python
#########################################################################
#                                                                       #    
# MY FUNCTION: in the function I set the vowel and consonant at 0, then #
#              with a first loop I find how many vowel substring and how#
#              many consonants substring are, in this way I can see you #
#              win with a if-else that prints the winner based on the   # 
#              number of vowel substrings or consonant substrings       #
#                                                                       #        
#########################################################################

def minion_game(string):
    vowel = 0
    consonant = 0
     
    for i in range(len(string)):
        if string[i] in 'AEIOU':
            vowel+=(len(string)-i)
        else:
            consonant+=(len(string)-i)
                
    if vowel < consonant:
        print('Stuart', consonant)
    elif vowel > consonant:
        print('Kevin', vowel)
    else:
        print('Draw')
    return

####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

if __name__ == '__main__':
    s = input()
    minion_game(s)
```
