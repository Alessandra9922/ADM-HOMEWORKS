### Exercise n° 5 - Insertion Sort - Part 1
##### Reference: https://www.hackerrank.com/challenges/insertionsort1/problem
##### This exercise want you to do a FUNCTION that given a sorted list with an unsorted number in the rightmost cell, insert it into the array

```python
####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'insertionSort1' function below.
#
# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER_ARRAY arr
#

########################################################################
#                                                                      #
# MY FUNCTION: Takes the last element of the array and it compares     # 
#               this element with previous elements, moving from right #
#                to left. If a previous element is larger than test,   #
#               it shifts that element to the right, once it finds the #
#               correct position for test (or reaches the start of the #
#               array), it inserts test there. After each shift or     #
#               insertion, it prints the current state of the array.   #
#               So I position the last element into its correct sorted # 
#               position in the array, printing each intermediate step.#
#                                                                      #
########################################################################

def insertionSort1(n, arr):
    test = arr[-1]
    for i in range(n-2, -1, -1):
        if arr[i] > test:
            arr[i+1] = arr[i]
            print(' '.join(map(str, arr)))
        else:
            arr[i+1] = test
            print(' '.join(map(str, arr)))
            return

    arr[0] = test
    print(' '.join(map(str, arr)))

####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    insertionSort1(n, arr)
```

### Exercise n° 6 - Insertion Sort - Part 2
##### Reference: https://www.hackerrank.com/challenges/insertionsort2/problem
##### This exercise want you to do a FUNCTION that inserts each element in an array at the right position so that the array is sorted.

```python
####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'insertionSort2' function below.
#
# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER_ARRAY arr
#

##########################################################################
#                                                                        #
# MY FUNCTION: It iteratively processes each element of the array,       #
#              starting from the second element. For each element,       #
#              it finds its correct position within the already sorted   #
#              portion of the array. It shifts larger elements to the    #
#              right to make space for test. Once the correct position   #
#              is found, it inserts test. I sorted the array step-by-step#
#                                                                        #
##########################################################################

def insertionSort2(n, arr):
    for i in range(1, n):
        test = arr[i]
        j = i - 1
        while j >= 0 and test < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = test
        print(' '.join(map(str, arr)))
        
####################
#                  #
#   CODE PROVIDED  #
#   BY HACKERANK   #
#                  #
####################

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    insertionSort2(n, arr)
```
